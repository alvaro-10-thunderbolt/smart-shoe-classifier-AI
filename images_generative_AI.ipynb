# model import



```python
from diffusers import StableDiffusionPipeline
import torch
```

#model load (small version)

```python
pipe=StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4",torch_dtype=torch.float16).to("cuda")
```

# image generate

```python
prompt="a futuristic smart shoe with AI desin, high detail, glowing effects"
image=pipe(prompt).images[0]

#show the image
image.show()

#save it
image.save("smart_shoe.png")
```
